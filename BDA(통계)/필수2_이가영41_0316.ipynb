### 필수과제 2
- pclass에 남,여까지 생존 확률 만들기
- 아마 6가지의 변수가 나올 것. 1.남 1여 2남 2여 3남 3여 만들기
- if, for로 만들기

```python
import seaborn as sns
import pandas as pd
import numpy as np
tt = sns.load_dataset('titanic')
tt
```

```python
# 정답이 아래와 같아야함
survival_prob = tt.groupby(['pclass', 'sex'])['survived'].mean()

print("pclass별 남,여까지 생존 확률:")
print(survival_prob)
```

```python
# 클래스별 생존 확률 계산
survival_prob = tt.groupby(['pclass', 'sex'])['survived'].mean()

print("pclass별 남,여까지 생존 확률:")
print(survival_prob)
```

```python
# 클래스별 여성과 남성의 수를 구함
woman3 = len(tt.query("pclass == 3 and sex == 'female'"))
man3 = len(tt.query("pclass == 3 and sex == 'male'"))
woman2 = len(tt.query("pclass == 2 and sex == 'female'"))
man2 = len(tt.query("pclass == 2 and sex == 'male'"))
woman1 = len(tt.query("pclass == 1 and sex == 'female'"))
man1 = len(tt.query("pclass == 1 and sex == 'male'"))

# 클래스별 생존 확률 계산
tt['pclass_prob'] = 0

for i in range(len(tt)):
    if tt.loc[i]['pclass'] == 3: # 3번 클래스로 조건 찾기
        if tt.loc[i]['sex'] == 'female':
            tt['pclass_prob'].loc[i] = woman3
        elif tt.loc[i]['sex'] == 'male':
            tt['pclass_prob'].loc[i] = man3
            
    elif tt.loc[i]['pclass'] == 2: # 2번 클래스의 조건 찾기 
        if tt.loc[i]['sex'] == 'female':
            tt['pclass_prob'].loc[i] = woman2
        elif tt.loc[i]['sex'] == 'male':
            tt['pclass_prob'].loc[i] = man2
        
    elif tt.loc[i]['pclass'] == 1: # 1번 클래스의 조건 찾기
        if tt.loc[i]['sex'] == 'female':
            tt['pclass_prob'].loc[i] = woman1
        elif tt.loc[i]['sex'] == 'male':
            tt['pclass_prob'].loc[i] = man1

# 생존 확률 계산
tt['pclass_prob'] = tt['survived'] / tt['pclass_prob']
```

```python
#죽으면 0, 살았으면 확률 부여
tt[['pclass','sex','pclass_prob']]
```
