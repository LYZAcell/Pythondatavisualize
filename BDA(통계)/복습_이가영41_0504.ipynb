### 중심 위치 척도
1. 산술평균
2. 중앙값(median)
3. 최빈값(frequent)


1. 산술평균
- 양적자료
- 평균은 자료를 더한 값을 그 자료의 수로 나눈 값
- 이상치(outlier)에 영향을 받는다. (도메인적으로도 판단)
- 생각보다 머신러닝, 데이터 통계치도 중요하지만 이런 기본 통계 이해도 매우 중요하다. 
- 가중평균: 중요도에 따라 측정한 값의 가중치를 고려한 평균 
    - 가중치 뿐만 아니라 거리 등에 따른 가중치에도 많이 사용한다. 
    
2. 중앙값
- 양적자료
- 중앙에 위치한 값. 짝수개라면 중앙에 이웃한 두 자료의 평균값
- 극단값에 영향을 받지 않음 -> 크기순으로 나열하기 때문
- 확장성이 없음
- 평균이 너무 왜곡됐을 때 차라리 median으로 사용하는 경우도 많음

3. 최빈값
- 범주형자료&양적자료 모두 적용 가능
- 범주형, 양적 자료에 모두 적용
- 최빈값은 없을수도, 여러개일수도 있다

## 정규화
##### feature이 늘어나면 평균은 위험할 수 있다

- 정규화: 데이터의 단위를 맞춰주면서 데이터의 분포를 같은 단위로 비교분석할 수 있게 해주는 작업
- 분포가 바뀌는게 아니라 같은 값으로 수치적 조정이 있을 뿐임 -> 정규화와 정규분포는 다른 용어임을 명심

1. 새로운 값 : 전값 / 최대값
- 상대 도수느낌. simple feature scaling 

2. min-max: Xnew = Xold-Xmin / Xmax - Xmin
- 새로운 값 = 오래된것 - 최소값 / 최대값 - 최소값
- 0 ~ 1사이의 값으로 바뀐다
- 이상치에 매우 민감하다(min,max라는 극단값으로 다루니까) 
- 때문에 이상치가 적고 데이터가 모든 페처에 대해 주우요도가 높을 때 사용하는 것이 좋다
- 신경망 입력 같은 입력 데이터 범위가 제한된 모델에서 성능이 좋을 수 있다

3. Z-score : Xnew = Xold-평균/표준편차
- 평균0, 표준편차가 1이도록 조정
    - 즉 0을 중심으로 퍼진다.
- 표준 정규분포를 만들어준다. 
- 흔히 말하는 대부분의 알고리즘은 정규분포를 가정한다 (특히 선형(linear)모델 같은 경우 정규분포로 짜진 데이터가 성능이 좋을 수 밖에 없다. )
- 이상치에 영향을 받지만 2에 대비 영향이 줄어든다. (이상치를 좀 줄여주긴한다. 다만 평균이 쓰이는 것의 한계)
- 정규분포를 가정하는 다양한 테스트, 머신러닝, 기타 분석에서 성능이 좋을 수 밖에 없다. 
- 정규 분포를 따르는 경우 Z-Score로 정규화해도 크게 문제되지는 않는다 -> 다양한 ML, 분석 시 정규분포는 성능이 좋을 수 밖에 없다.


## 도수분포표
- 각 계급에 속하는 도수를 조사하여 나타낸 표
- 계급을 정하고 ( = 연령대 ) , 계급값(연령대의 중앙값), 도수(연령에 속하는 인원)
    - 이를 나눠서 도수분포표를 작성한다 ( 각 계급에 속하는 도수를 조사하여 나타낸 표)
    - 각 연령대에 속하는 인원을 조사하여 나타낸 표
    

## 막대그래프와 히스토그램의 차이 + Boxplot
- 막대그래프: 범주로 그분. 막대간 간격 유지 (ex: 김치찌개, 비빔밥, 돈까스 등) 
- 히스토그램: 측정된 연속적인 값(몸무게, 키, 성적 등) 표현. 순서를 임의로 바꿀 수 없다
    - bins 개수 조정할 수록 그래프가 왜곡 될 수 있는 것 (보통 15~20개 정도 둔다) 
    
## Boxplot
- 동그라미는 아웃라이어
- 점선은 상위분위수에서 1.5지점 (잘 본적 없다)
- 선의 맨 끝:  상위 경계 내 측의 최대값
- 상자 맨위: 삼분위(75%)
- 상자 안 선 : 중위값(median)
- 상자 맨아래 : 일분위(25%)
- 맨아래 선: 하위 경계 내 최소값
- 하위 경계 :하위 삼분위수에서 1.5*IQR아래지점
- 4분위수 범위를 IQR이라함. 점선은 사분위 선 위/아래에 1.5 곱한 지점
    - IQR : Q3-Q1. 여기서 1.5 곱한거까지로 정상 범위로 취급하고 그 이상은 이상치로 취급
    - 여기서는 1.5곱했지만 이게 default는 아니다. 분포마다 구간 비교하면서 2를곱하기도 5를 곱하기도하면서 조정한다
    - 다만 1.5는 어느정도 증명된 수라 보편적으로 사용될 뿐
    
#### 데이터에서 아웃라이어가 나오면 어떻게해야할까?

- 제외시키거나 대표값으로 쓰는 방법
- 사실 무작정 지우면 안된다 : 강사님의 경우 지우지 않음. 
    - 데이터에서 통계로 너무 치우치면 또 수치싸움이 된다 -> 이러면 도메인(비즈니스)적 헛점이 될 수 있다
    - 아웃라이어라고 고객에서 제외하면 1억쓴 VIP일 수도 있잖아
- 값 자체에 의미가 없다면(논리가 없다면) 제거해도되지만
- 도메인적으로 고찰했을 때 의미가 있는사람이라면 (VIP는 적지만 수요가 엄청남) 단순히 지우면 안된다
    - 의미하는 바를 바라보고 아웃라이어만 오히려 빼서 도출하거나 평균과 아웃라이어간의 가중치를 줘서 패턴을 줄 수도 있다
- 아웃라이어를 지워도지워도 계속 아웃라이어가 나올 수 있다 -> 그렇게 다 거른 자료를 일반화할 수 있냐에 대한 고민이 필요하다. 
- 비즈니스, 도메인적 고찰이 필요한 파트다. 

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# titanic으로 살펴보기

tt = sns.load_dataset('titanic')
tt
```

```python
tt_sp = tt[['age','fare']] 
tt_sp
```

```python
# 결측치 채우기
tt_sp['age'].fillna(tt_sp['age'].median(),inplace = True)
tt_sp['fare'].fillna(tt_sp['fare'].median(),inplace = True)
```

```python
#결측치 대체 확인
tt_sp.isna().sum()
```

```python
#Min
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
tt_sp['age_mm'] = min_max_scaler.fit_transform(tt_sp[['age']])
tt_sp['fare_mm'] = min_max_scaler.fit_transform(tt_sp[['fare']])
```

```python
#z-score
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
tt_sp['age_z'] = standard_scaler.fit_transform(np.array(tt_sp['age']).reshape(-1,1))
tt_sp['fare_z'] = standard_scaler.fit_transform(np.array(tt_sp['fare']).reshape(-1,1))
```

```python
# 코드올라오면 참고하기 
# 핵심은 fare은 정규분포가 아니라는 점. 

# 무작위 데이터 생성
np.random.seed(0)
data = np.random.randn(1000) * 100 + 20

# 데이터 프레임 생성
df = pd.DataFrame(data, columns=['Original'])

# Min-Max Scaling
scaler_min_max = MinMaxScaler()
df['Min-Max Scaled'] = scaler_min_max.fit_transform(df[['Original']])

# Z-Score Normalization
scaler_standard = StandardScaler()
df['Z-Score Scaled'] = scaler_standard.fit_transform(df[['Original']])

# 시각화
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# 원본 데이터 히스토그램
sns.histplot(df['Original'], bins=30, kde=True, ax=ax[0])
ax[0].set_title('Histogram of Original Data')

# Min-Max Scaled 데이터 히스토그램
sns.histplot(df['Min-Max Scaled'], bins=30, kde=True, ax=ax[1], color='green')
ax[1].set_title('Histogram of Min-Max Scaled Data')

# Z-Score Scaled 데이터 히스토그램
sns.histplot(df['Z-Score Scaled'], bins=30, kde=True, ax=ax[2], color='red')
ax[2].set_title('Histogram of Z-Score Scaled Data')

plt.tight_layout()
plt.show()

# 박스 플롯으로 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Box Plot of Original, Min-Max and Z-Score Scaled Data')
plt.show()
```
