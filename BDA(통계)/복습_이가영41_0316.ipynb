## 기존에 배운 Python 문법을 이용해서 for , if 문으로 다양한 파생변수를 만들어 보자!

- for , if 다양하게 파생변수 만들어 보자!
-  pclass 별로 1,2,3 등급 , survived 확률을 만들어보자!

```python
# 둘 차이: tt[pclass]는 불리언으로 처리된 모든 값 출력. 쿼리로 줘야 3등급인 애들이 잘 출력된다.
print(len(tt['pclass']==3))
print(len(tt.query('pclass==3')))
```

```python
tt['pclass']
```

```python
tt.loc[0]
```

```python
#파생변수 컬럼을 만들어야 한다.
# 각 클래스 총원 확인 후 분모는 기재한 것. 생존 확률을 각 클래스마다 정리해서 넣어준다.
tt['pclass_prob'] =0

for i in range(len(tt)):
    if tt.loc[i]['pclass'] == 3: # 3번 클래스로 조건 찾기
        tt['pclass_prob'].loc[i]= tt.loc[i]['survived'].sum()/491 # 3번 클래스의 생존확률
    elif tt.loc[i]['pclass'] ==2: # 2번 클래스의 조건 찾기 
        tt['pclass_prob'].loc[i] = tt.loc[i]['survived'].sum()/184 # 2번 클래스의 생존확률
    elif tt.loc[i]['pclass'] ==1: # 1번 클래스의 조건 찾기
        tt['pclass_prob'].loc[i] = tt.loc[i]['survived'].sum()/216 # 1번의 클래스의 생존확률
```

```python
#죽으면 0, 살았으면 확률 부여
tt[['pclass','pclass_prob']]
```

```python
# 2개 이상인 경우는 index, 컬럼을 지정해서 crosstab을 만들 수 있다.
pd.crosstab(index = [tt['sex'],tt['pclass']],columns = tt['survived'])
```

```python
pd.crosstab(index = [tt['age'],tt['pclass']],columns = tt['sex'])
```

```python
# 전체 sum 도 보면 좋다.
# 인자값에 넣기만 하면 된다.
# 인자 값은 margins , 확률은 normailize  True, False

# 2개 이상인 경우는 index, 컬럼을 지정해서 crosstab을 만들 수 있다.
# 크로스탭으로 간단하게 다양한 컬럼들간의 관계, 확률표도 만들 수 있다.
pd.crosstab(index = [tt['sex'],tt['pclass']],columns = tt['survived'], margins = True, normalize=True)
```

###  assign() 파생변수 만들기
- assign(파생변수명 = 요구조건등을 적는 것)

- numpy에서 제공하는 where 함수를 이용해서 파생변수를 만들 것. 원래 반복문으로 돌려야하는 것을 넘파이가 알아서 돌며 인덱스를 반환해주는 특징

```python
tt.assign(age_old= np.where(tt['age']>30, 'old','young')) # nan값들은 그외로 처리된다.
```

```python
tt_old = tt.assign(age_old= np.where(tt['age']>30, 'old','young')) # nan값들은 그외로 처리된다.
# age_old라는 파생변수에 old, young 여부가 들어옴
tt_old[['survived','pclass','age','age_old']]
```

```python
# 운임별 rich, poor 
tt_rich = tt.assign(tt_rich = np.where(tt['fare']>50, 'rich', 'poor'))
tt_rich
```

```python
np.where(tt['age']>0) # 안의 조건에 들어간 조건들의 인덱스를 반환한다.
```

```python
# 넘파이가 조건의 인덱스를 반환해줌. tt['age']인덱스를 가져와 old,young을 맵핑해주는 것. 
tt.assign(age_old = np.where(tt['age']>30, 'old','young')) #30세 넘으면 old 처리
```
