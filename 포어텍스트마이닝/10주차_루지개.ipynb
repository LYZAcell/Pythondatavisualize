```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./brio_totals - Sheet1.csv")
df
```

```python
plt.figure(figsize=(10, 6))  # 그래프 크기 설정 (선택 사항)

plt.bar(df['state'], df['vaccination'])  # 막대 그래프 생성

plt.title('Vaccination by State')  # 그래프 제목
plt.xlabel('state')  # x축 레이블
plt.ylabel('vaccination')  # y축 레이블

plt.show()  # 그래프 표시
```
