````python
# import pandas as pd  # pd.read_excel의 의존성이 openpyxl임
import openpyxl #활용해서 csv를 xlsx로 다루기 가능
````

````python
score = pd.read_excel('score_1.xlsx')
score
````

````python
score.index = ['H','K','SU']
score
````

````python
score.columns =['name','eng','math','kor','soc']
score
````

````python
# set index문법: 해당 컬럼으로 index가 될 수 있게함
score = score.set_index('name')
score
````

````python
# reset_index : 다시 디폴트인덱스로 돌아감
# set, reset 모두 변수에 함수를 할당해줘야한다.
score = score.reset_index()
score
````

- 컬럼 추가하기

````python
score ['por'] = [50,45,40]
score
````

- copy 메소드

## loc, iloc : 행가져오기

- loc은 location

- df.loc['인덱스명칭'] : 인덱스 이름으로 가져오기
  - ex: df.loc[10,'content']

- df.iloc[인덱스번호]: 인덱스 번호로 가져오기
  - 인덱스를 컷하면 상대적으로 할당되게 된다.

````python
score = score.reset_index()
score
````

````python
score.loc[1,'eng']
````

````python
score.iloc[0]
````
